```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
existing_df = pd.read_csv('breast-cancer-data.csv',index_col = 0, thousands = ',')
existing_df.head()
```

```python
existing_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(existing_df['diagnosis'])
existing_df['diagnosis']=le.transform(existing_df['diagnosis'])
existing_df.head()
```

```python
x_temp = existing_df.iloc[:,1:]
x_temp.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(x_temp)
```

```python
x_temp_2d = pca.transform(x_temp)
x_temp_2d = pd.DataFrame(x_temp_2d)
x_temp_2d.index = x_temp.index
x_temp_2d.columns = ['PC1','PC2']
x_temp_2d.head()
```

```python
pca.explained_variance_ratio_
```

```python
new_df = pd.concat([existing_df.iloc[:,0],x_temp_2d],axis=1)
new_df.head()
```

Now calculating the efficiency before and after decomposition

#1. Before decomposition

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = existing_df.iloc[:,1:].values
y = existing_df.iloc[:,0].values
```

```python
X_train,x_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=0)
```

```python
log = LogisticRegression(random_state=0)
log.fit(X_train,y_train)
y_pred = log.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred, normalize=True)
```

2. Now after decomposition

```python
X = new_df.iloc[:,1:].values
y = new_df.iloc[:,0].values
```

```python
X_train,x_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=0)
```

```python
log = LogisticRegression(random_state=0)
log.fit(X_train,y_train)
y_pred = log.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred, normalize=True)
```
